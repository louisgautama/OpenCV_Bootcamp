```python
import cv2
import easyocr
from IPython.display import Image
```

```python
Image("/content/resume.png")
```

```python
reader = easyocr.Reader(["en"])
```

```python
output = reader.readtext("/content/resume.png")
```

```python
output
```

```python
words = [item[1] for item in output]
print(words)
```

```python
with open('readme.txt', 'a') as f:
    f.write('\n'.join(words))
```
